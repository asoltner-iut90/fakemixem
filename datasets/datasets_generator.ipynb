```python
from youtubeTools.youtube_fetcher import *
import time
import csv
from tqdm import tqdm
import pandas as pd
```

```python
channel = "https://www.youtube.com/sheshounet"
```

```python
video_urls = get_all_channel_videos(channel)
print(f"Number of videos found: {len(video_urls)}")
```

```python
videos = []
```

```python
channel_csv = "amixem_20251022.csv"

def load_csv(channel_csv):
    with open(channel_csv, newline="", encoding="utf-8") as f:
        reader = csv.DictReader(f)
        data = list(reader)
    return data


videos = load_csv(channel_csv)
```

```python
def video_already_fetched(video_url, videos):
    for video in videos:
        if video["webpage_url"] == video_url:
            return True
    return False
```

```python
best_keys = [
    "webpage_url", "id", "title", "thumbnail", "description", "channel_id", "duration", "view_count",
    "age_limit", "age_limit", "categories", "tags", "upload_date", "timestamp", "fulltitle", "is_live","was_live", "comment_count"
]

errored_video_urls = []

for i in tqdm(range(len(video_urls)), desc="Traitement en cours"):
    url = video_urls[i]
    try:
        if video_already_fetched(url, videos):
            continue
        video_info = get_video_info(url)
        filtered_info = {key: video_info.get(key) for key in best_keys}
        videos.append(filtered_info)
        time.sleep(1)
    except Exception as e:
        errored_video_urls.append(url)
        print(f"Error fetching video {url}: {e}")

print(f"Number of videos successfully fetched: {len(videos)}")
print(f"Number of videos errored: {len(errored_video_urls)}")
```

```python
date = time.strftime("%Y%m%d")
file = f"amixem_{date}.csv"

def save_to_csv(data, file):
    with open(file, "w", newline="", encoding="utf-8") as f:
        writer = csv.DictWriter(f, fieldnames=data[0].keys())
        writer.writeheader()
        writer.writerows(data)

save_to_csv(videos, file)
```

```python
for video in tqdm(videos, desc="Fetching likes and dislikes"):
    if "like" not in video:
        try:
            video_info = get_dislikes(video["id"])
            video["likes"] = video_info["likes"]
            video["dislikes"] = video_info["dislikes"]
            time.sleep(0.5)
        except Exception as e:
            video["likes"] = None
            video["dislikes"] = None
            print(f"Error fetching likes for video {video['webpage_url']}: {e}")
```

```python
for video in tqdm(videos, desc="Fetching if the video is sponsored"):
    if "sponsored" not in video:
        try:
            sponsored = is_sponsored(video["id"])
            video["sponsored"] = sponsored
            time.sleep(0.5)
        except Exception as e:
            video["sponsored"] = None
            print(f"Error fetching sponsorship for video {video['webpage_url']}: {e}")
```
