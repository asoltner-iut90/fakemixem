```python
import pandas
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../Datasets/amixem_20251023.csv")
df.head()
```

```python
df['upload_date'] = pd.to_datetime(df['upload_date'], format='%Y%m%d')
df.head()
```

```python
df["precise_upload_date"] = pd.to_datetime(df['timestamp'], unit='s')
df.head()
```

```python
# graph of the advancement of the number of views over time
df = df.sort_values(by='upload_date')
plt.figure(figsize=(12, 6))
plt.plot(df['upload_date'], df['view_count'], marker='o')
plt.title("Number of Views Over Time")
plt.xlabel("Upload Date")
plt.ylabel("View Count")
plt.show()
```

```python
# graph of the number of videos uploaded per year
data = df['upload_date'].dt.year.value_counts().sort_index()
plt.figure(figsize=(10, 5))
data.plot(kind='bar')
plt.title("Number of Videos Uploaded Per Year")
plt.xlabel("Year")
plt.ylabel("Number of Videos")
plt.show()
```
